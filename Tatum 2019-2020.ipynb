```python
from nba_api.stats.endpoints import playergamelog
from nba_api.stats.static import players

wood_id = players.find_players_by_full_name("Christian Wood")[0]['id']
hern_id = players.find_players_by_full_name("Willy Hernangomez")[0]['id']

wood_games = playergamelog.PlayerGameLog(player_id=wood_id)
wood_games = wood_games.get_data_frames()[0]

hern_games = playergamelog.PlayerGameLog(player_id=hern_id)
hern_games = hern_games.get_data_frames()[0]

hern_games.head()
```

```python
import pandas as pd

raptor = pd.read_csv("modern_RAPTOR_by_player.csv")
raptor.head()

highwar = raptor.loc[raptor['war_total'] > 20 ]
highwar.head()
```

```python
import matplotlib.pyplot as plt


raptor_plot = highwar.plot(kind="scatter", x = 'poss', y = 'predator_total', grid = 1)

for i, point in highwar.iterrows():
    raptor_plot.text(point['poss'] +50, point['predator_total']+.02, str(point['player_name'] +' ' + str(point['season'])))
        
        
```
